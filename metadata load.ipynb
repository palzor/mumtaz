Import the framework library first

```
import mumtaz as mt
```

Pass the Azure datafactory pipline name as this would help us identify the records

```
dbutils.widgets.text('ADFPipelineName',"None")
pipelineName=dbutils.widgets.get("ADFPipelineName")
```

Create the object of the class DBFrameworkHelper, in the example below we are creating the object named MDF

```
MDF=mt.DBFrameworkHelper(None if pipelineName == "None" else pipelineName)
```

If you want to cleanup the metadata database then use cleanup function 
MDF.cleanup()
and then create the database using 
MDF.init()
You can skip if its already created

```
MDF.pipelinename
```

Please make sure that the folders are mounted. if not then you could use <b>MDF.mount()</b> to mount the folders 
<br>Mount function accepts following parameters:
<br>
<ol>
  <li>Containername</li>
  <li>Storageaccountname</li>
  <li>scopename</li>
  <li>keyname</li>
</ol>
<br>

```
MDF.mountFolder('sourceschema','bnz101sa','KeyVaulScope','storageaccountkey')
```

```
MDF.mountFolder('landing','bnz101sa','KeyVaulScope','storageaccountkey')
```

```
MDF.mountFolder('dbfs','bnz101sa','KeyVaulScope','storageaccountkey')
```

Once we have the folder available then we need to run the fucions to load the metadata and the data in the foloowing order:

<ol> 
  <br>
  <ol>Creating the metadata
    <li>MDF.init() </li>
    <li>MDF.process_metadata_excludedColumns() </li>
    <li>MDF.process_metadata_tableList() </li>
    <li>MDF.process_metadata_loadSourceSchema()  </li>
    <li>MDF.process_metadata_loadSourceConstraints() </li>
    <li>MDF.process_metadata_columns() </li>
    <li>MDF.process_metadata_tables() </li>
    <li>MDF.updateMetadataTables() </li>
  </ol>
  <br>
  <ol>Loading the data
    <li>MDF.execute_load_data()</li>
  </ol>
</ol>
